```python
#@title
%%manim -qm -v WARNING SnellsLov

class SnellsLov(Scene):
  def construct(self):

    t = ValueTracker(0)
    P = Dot(point=[-2,3,0])
    P_tekst = Tex("P").next_to(P,UP+LEFT)
    StartPunkt = VGroup(P,P_tekst)

    Q = Dot(point=[3,-3,0])
    Q_tekst = Tex("Q").next_to(Q,DOWN+RIGHT)
    SlutPunkt = VGroup(Q,Q_tekst)

    overgang = always_redraw(
        lambda: DashedLine(start=[-3,t.get_value(),0],end=[3,t.get_value(),0])
    )
    a = always_redraw(
        lambda: Dot(point=[0,t.get_value(),0])
    )
    StrleEt = always_redraw(
        lambda: Line(start=P,end=a,color=GOLD)
    )
    StrleTo = always_redraw(
        lambda: Line(start=a,end=Q,color=GOLD)
    )


    self.add(overgang,a,StartPunkt,SlutPunkt,StrleEt,StrleTo)
    self.play(
        t.animate.set_value(2),
        run_time=2
    )
    self.play(
        t.animate.set_value(-1.5),
        run_time=2
    )
    self.play(
        t.animate.set_value(0),
        run_time=2
    )
    self.wait()
    
```

```python
%%manim -qk -v WARNING SnellsLovTo

class SnellsLovTo(Scene):
  def construct(self):

    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))

    t = ValueTracker(3)
    P = Dot(point=[0,3,0])
    P_tekst = Tex("P", tex_template=myTemplate).next_to(P,UP+LEFT)
    StartPunkt = VGroup(P,P_tekst)

    Q = Dot(point=[5,-2,0])
    Q_tekst = Tex("Q", tex_template=myTemplate).next_to(Q,DOWN+RIGHT)
    SlutPunkt = VGroup(Q,Q_tekst)

    #overgang = always_redraw(
    #    lambda: DashedLine(start=[-3,t.get_value(),0],end=[3,t.get_value(),0])
    #)
    grnseflade = DashedLine(start=[-1,0,0],end=[6,0,0],color=BLUE)
    a = always_redraw(
        lambda: Dot(point=[t.get_value(),0,0])
    )
    a_tekst=Tex("a", tex_template=myTemplate).next_to(a,UP)

    Strlegang = Line(start=P.get_center(),end=Q.get_center(),color=GOLD)
    StrleEt = always_redraw(
        lambda: Line(start=P.get_center(),end=[t.get_value(),0,0],color=GOLD)
    )
    StrleTo = always_redraw(
        lambda: Line(start=[t.get_value(),0,0],end=Q.get_center(),color=GOLD)
    )

    b  = BraceBetweenPoints([5,0,0],[5,-2,0],direction=RIGHT,stroke_width=1)
    c  = BraceBetweenPoints([0,3,0],[0,0,0],stroke_width=1)
    L  = BraceBetweenPoints([0,-2,0],[5,-2,0],stroke_width=1)
    x  = BraceBetweenPoints([0,0,0],[2,0,0],stroke_width=1)
    Lx = BraceBetweenPoints([2,0,0],[5,0,0],direction=UP,stroke_width=1)
    b_tekst  = Tex("b", tex_template=myTemplate).next_to(b,RIGHT)
    c_tekst  = Tex("c", tex_template=myTemplate).next_to(c,LEFT)
    L_tekst  = Tex("L", tex_template=myTemplate).next_to(L,DOWN)
    x_tekst  = Tex("x", tex_template=myTemplate).next_to(x,DOWN)
    Lx_tekst = Tex("L-x", tex_template=myTemplate).next_to(Lx,UP)
    lngder = VGroup(b,b_tekst,c,c_tekst,L,L_tekst,x,x_tekst,Lx,Lx_tekst)

    versteTrekant = Polygon([0,3,0],[0,0,0],[2,0,0],color=BLUE_C,fill_opacity=0.5)
    NedersteTrekant = Polygon([2,0,0],[5,0,0],[5,-2,0],color=BLUE_C,fill_opacity=0.5)

    alpha = Angle(Line(start=[0,0,0],end=[0,4,0]),Line(start=[-2,6,0],end=[4,-3,0]), quadrant=(-1,1))
    beta  = Angle(Line(start=[-1,2,0],end=[8,-4,0]),Line(start=[5,-3,0],end=[5,0,0]), quadrant=(-1,1), other_angle=True)
    alpha_tekst = MathTex(r"\bm \upalpha", tex_template=myTemplate).next_to(alpha.point_from_proportion(0.5),DR,buff=SMALL_BUFF)#.next_to(alpha,DOWN+0.005*RIGHT,buff=0.1)
    beta_tekst = MathTex(r"\bm \upbeta", tex_template=myTemplate).next_to(beta.point_from_proportion(0.5),UL,buff=SMALL_BUFF)#.next_to(beta,UP+0.005*LEFT,buff=0.1)
    vinkel_alpha = VGroup(alpha,alpha_tekst)
    vinkel_beta = VGroup(beta,beta_tekst)

    indfaldslod = DashedLine(start=[2,1.5,0],end=[2,-1.5,0],color=GREEN)

    indfaldsvinkel = Angle(indfaldslod, Line(start=[-2,6,0],end=[4,-3,0]), quadrant=(-1,-1))
    brydningsvinkel = Angle(Line(start=[-1,2,0],end=[8,-4,0]),indfaldslod,quadrant=(1,1),other_angle=True)
    indfaldsvinkel_tekst = MathTex(r"i", tex_template=myTemplate).next_to(indfaldsvinkel,UP+0.005*LEFT,buff=0.1)
    brydningsvinkel_tekst = MathTex(r"b", tex_template=myTemplate).next_to(brydningsvinkel,DOWN+0.005*RIGHT,buff=0.1)
    IndfaldsvinkelTotal = VGroup(indfaldsvinkel,indfaldsvinkel_tekst)
    BrydningsvinkelTotal = VGroup(brydningsvinkel,brydningsvinkel_tekst)    

    self.add(StartPunkt,SlutPunkt)
    self.wait(2)
    self.bring_to_front(StartPunkt,SlutPunkt)
    self.play(Create(Strlegang))
    self.bring_to_front(StartPunkt,SlutPunkt)
    self.wait(2)
    self.play(Create(grnseflade))
    self.bring_to_front(StartPunkt,SlutPunkt)
    self.wait(2)
    self.add(StrleEt,StrleTo)
    self.bring_to_front(StartPunkt,SlutPunkt)
    self.remove(Strlegang)
    self.add(a)
    self.wait(2)
    
    self.play(
        t.animate.set_value(4),
        MaintainPositionRelativeTo(a_tekst,a),
        run_time=4,
        
        #rate_func=rate_functions.linear
    )
    self.play(
        t.animate.set_value(0.5),
        MaintainPositionRelativeTo(a_tekst,a),
        run_time=4,
        #rate_func=rate_functions.linear
    )
    self.play(
        t.animate.set_value(2),
        MaintainPositionRelativeTo(a_tekst,a),
        run_time=4,
        #rate_func=rate_functions.linear
    )
    self.wait(2)

    #self.bring_to_back(versteTrekant)
    #self.play(Create(versteTrekant))
    #self.wait()
    #self.play(Create(c),Create(x),Write(c_tekst),Write(x_tekst))
    
    #Den indkomne strle
    self.play(Create(c),Write(c_tekst))
    self.wait(2)
    self.play(Create(x),Write(x_tekst))
    self.wait(2)

    #Den brudte strle
    self.play(Create(b),Write(b_tekst))
    self.wait(2)
    self.play(Create(L),Write(L_tekst))
    self.wait(2)
    self.play(Create(Lx),Write(Lx_tekst))
    self.wait(2)
    
    lign1 = MathTex(r"|Pa|=",r"\sqrt{x^2+c^2}", tex_template=myTemplate).scale(0.5).to_corner(UL)
    lign2 = MathTex(r"|aQ|=",r"\sqrt{(L-x)^2+b^2}", tex_template=myTemplate).scale(0.5).next_to(lign1,DOWN).align_to(lign1,LEFT)
    lign3a = MathTex(r"t=",r"{|Pa|",r"\over",r"v_1",r"}+",r"{|aQ|",r"\over",r"v_2",r"}", tex_template=myTemplate).scale(0.5).next_to(lign2,DOWN).align_to(lign1,LEFT)
    lign3b = MathTex(r"t=",r"{\sqrt{x^2+c^2}",r"\over",r"v_1",r"}+",r"{\sqrt{(L-x)^2+b^2}",r"\over",r"v_2",r"}", tex_template=myTemplate).scale(0.5).next_to(lign2,DOWN).align_to(lign1,LEFT)
    lign4a = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"1",r"\over 2\sqrt{x^2+c^2}}\cdot",r"2x",r"+{1\over v_2}\cdot{",r"1",r"\over 2\sqrt{(L-x))^2+b^2}}",r"\cdot",r"2\cdot(L-x)",r"\cdot",r"(-1)", tex_template=myTemplate).scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4b = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"2x",r"\over 2\sqrt{x^2+c^2}}",r"-{1\over v_2}\cdot{",r"2\cdot(L-x)",r"\over 2\sqrt{(L-x))^2+b^2}}", tex_template=myTemplate).scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4c = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"x",r"\over \sqrt{x^2+c^2}}",r"-{1\over v_2}\cdot",r"{(L-x)\over \sqrt{(L-x))^2+b^2}}", tex_template=myTemplate).scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4d = MathTex(r"{dt\over dx}={1\over v_1}\cdot",r"sin(\alpha)",r"-{1\over v_2}\cdot",r"sin(\beta)", tex_template=myTemplate).scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign5 = MathTex(r"{dt\over dx}=0", tex_template=myTemplate).scale(0.5).next_to(lign4a,DOWN).align_to(lign1,LEFT)
    lign6a = MathTex(r"{1",r"\over",r"v_1",r"}\cdot",r"sin(\alpha)",r"-{",r"1",r"\over",r"v_2",r"}\cdot",r"sin(\beta)",r"=0", tex_template=myTemplate).scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6b = MathTex(r"{1",r"\over",r"v_1",r"}\cdot",r"sin(\alpha)",r"={",r"1",r"\over",r"v_2",r"}\cdot",r"sin(\beta)", tex_template=myTemplate).scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6c = MathTex(r"{",r"sin(\alpha)",r"\over",r"sin(\beta)",r"}",r"=",r"{",r"v_1",r"\over",r"v_2",r"}", tex_template=myTemplate).scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6d = MathTex(r"{sin(i)",r"\over",r"sin(b)",r"}",r"=",r"{",r"v_1",r"\over",r"v_2}", tex_template=myTemplate).scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    

    #trekanter
    #self.play(Create(versteTrekant),Create(NedersteTrekant))
    self.play(Create(versteTrekant))
    self.bring_to_back(versteTrekant)
    self.wait(2)
    self.play(Write(lign1))
    self.wait(2)

    self.play(Create(NedersteTrekant))
    self.bring_to_back(NedersteTrekant)
    self.wait(2)
    self.play(Write(lign2))
    self.wait(2)

    self.play(Write(lign3a))
    self.wait(2)

    self.play(Indicate(lign3a[1]))
    self.wait(2)
    self.play(Indicate(lign1[1]))
    self.wait(2)
    self.play(Indicate(lign3a[5]))
    self.wait(2)
    self.play(Indicate(lign2[1]))
    self.wait(2)

    self.play(Transform(lign3a,lign3b))
    self.wait(2)
    
    self.play(
            *[FadeOut(mob)for mob in lngder],
            lag_ratio=0.5,
            run_time=2
        )
    self.play(FadeOut(versteTrekant),FadeOut(NedersteTrekant))

    self.play(Write(lign4a))
    self.wait(2)
    
    self.play(TransformMatchingTex(lign4a,lign4b),run_time=2)
    self.wait(2)
    self.play(TransformMatchingTex(lign4b,lign4c),run_time=2)
    self.wait(2)

    self.play(
            *[FadeIn(mob)for mob in lngder],
            lag_ratio=0.5,
            run_time=2
        )
    self.play(FadeIn(versteTrekant),FadeIn(NedersteTrekant))
    self.bring_to_back(versteTrekant,NedersteTrekant)
    
    #Den brudte strle
    

    self.play(Create(vinkel_alpha))
    self.wait(2)
    self.play(Create(vinkel_beta))
    self.wait(2)

    self.play(TransformMatchingTex(lign4c,lign4d),run_time=2)
    self.wait(2)
    self.play(Write(lign5))
    self.wait(2)
    self.play(Write(lign6a))
    self.wait(2)
    self.play(TransformMatchingTex(lign6a,lign6b),run_time=2)
    self.wait(2)
    self.play(TransformMatchingTex(lign6b,lign6c),run_time=2)
    self.wait(2)
    

    self.play(
            #*[Uncreate(mob) for mob in lngder],
            *[FadeOut(mob) for mob in lngder],
            lag_ratio=0.5,
            run_time=2
        )
    self.play(FadeOut(a_tekst))
    self.wait(2)

    self.play(Create(indfaldslod))
    self.wait(2)

    self.play(Transform(vinkel_alpha,IndfaldsvinkelTotal))
    self.wait(2)
    self.play(Transform(vinkel_beta,BrydningsvinkelTotal))
    self.wait(2)

    self.play(TransformMatchingTex(lign6c,lign6d),run_time=2)
    self.wait(2)
    

    self.play(
            *[FadeOut(mob)for mob in self.mobjects],
            lag_ratio = 0.25,
            run_time=2
        )
    self.wait(5)


    #self.add(b,c,x,L,Lx)
    #self.add(b_tekst,c_tekst,x_tekst,L_tekst,Lx_tekst)
    self.wait()
    
```

```python
%%manim -qm -v WARNING Ligninger

class Ligninger(Scene):
  def construct(self):
    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    
    lign1 = MathTex(r"|Pa|=",r"\sqrt{x^2+c^2}").scale(0.5).to_corner(UL)
    lign2 = MathTex(r"|aQ|=",r"\sqrt{(L-x)^2+b^2}").scale(0.5).next_to(lign1,DOWN).align_to(lign1,LEFT)
    lign3a = MathTex(r"t=",r"{|Pa|",r"\over",r"v_1",r"}+",r"{|aQ|",r"\over",r"v_2",r"}").scale(0.5).next_to(lign2,DOWN).align_to(lign1,LEFT)
    lign3b = MathTex(r"t=",r"{\sqrt{x^2+c^2}",r"\over",r"v_1",r"}+",r"{\sqrt{(L-x)^2+b^2}",r"\over",r"v_2",r"}").scale(0.5).next_to(lign2,DOWN).align_to(lign1,LEFT)
    #lign4a = MathTex(r"{dt\over dx}=",r"{1\over v_1}",r"\cdot",r"{1\over 2\sqrt{x^2+c^2}}",r"\cdot",r"2x",r"+",r"{1\over v_2}",r"\cdot",r"{1\over 2\sqrt{(L-x))^2+b^2}}",r"\cdot",r"2\cdot(L-x)",r"\cdot",r"(-1)").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    #lign4b = MathTex(r"{dt\over dx}=",r"{1\over v_1}",r"\cdot",r"{2x\over 2\sqrt{x^2+c^2}}",r"-",r"{1\over v_2}",r"\cdot",r"{2\cdot(L-x)\over 2\sqrt{(L-x))^2+b^2}}").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    #lign4c = MathTex(r"{dt\over dx}=",r"{1\over v_1}",r"\cdot",r"{x\over \sqrt{x^2+c^2}}",r"-",r"{1\over v_2}",r"\cdot",r"{(L-x)\over \sqrt{(L-x))^2+b^2}}").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    #lign4d = MathTex(r"{dt\over dx}=",r"{1\over v_1}",r"\cdot",r"sin(\alpha)",r"-",r"{1\over v_2}",r"\cdot",r"sin(\beta)").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4a = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"1",r"\over 2\sqrt{x^2+c^2}}\cdot",r"2x",r"+{1\over v_2}\cdot{",r"1",r"\over 2\sqrt{(L-x))^2+b^2}}",r"\cdot",r"2\cdot(L-x)",r"\cdot",r"(-1)").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4b = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"2x",r"\over 2\sqrt{x^2+c^2}}",r"-{1\over v_2}\cdot{",r"2\cdot(L-x)",r"\over 2\sqrt{(L-x))^2+b^2}}").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4c = MathTex(r"{dt\over dx}={1\over v_1}\cdot{",r"x",r"\over \sqrt{x^2+c^2}}",r"-{1\over v_2}\cdot",r"{(L-x)\over \sqrt{(L-x))^2+b^2}}").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign4d = MathTex(r"{dt\over dx}={1\over v_1}\cdot",r"sin(\alpha)",r"-{1\over v_2}\cdot",r"sin(\beta)").scale(0.5).next_to(lign3a,DOWN).align_to(lign1,LEFT)
    lign5 = MathTex(r"{dt\over dx}=0").scale(0.5).next_to(lign4a,DOWN).align_to(lign1,LEFT)
    lign6a = MathTex(r"{1",r"\over",r"v_1",r"}\cdot",r"sin(\alpha)",r"-{",r"1",r"\over",r"v_2",r"}\cdot",r"sin(\beta)",r"=0").scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6b = MathTex(r"{1",r"\over",r"v_1",r"}\cdot",r"sin(\alpha)",r"={",r"1",r"\over",r"v_2",r"}\cdot",r"sin(\beta)").scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6c = MathTex(r"{",r"sin(\alpha)",r"\over",r"sin(\beta)",r"}",r"=",r"{",r"v_1",r"\over",r"v_2",r"}").scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    lign6d = MathTex(r"{sin(i)",r"\over",r"sin(b)",r"}",r"=",r"{",r"v_1",r"\over",r"v_2}").scale(0.5).next_to(lign5,DOWN).align_to(lign1,LEFT)
    

    self.play(Write(lign1))
    self.wait()
    self.play(Write(lign2))
    self.wait()
    self.play(Write(lign3a))
    self.wait()

    self.play(Transform(lign3a,lign3b))
    self.wait()
    self.play(Write(lign4a))
    self.wait()
    #self.play(ReplacementTransform(lign4a,lign4b))
    self.play(TransformMatchingTex(lign4a,lign4b),run_time=2)
    self.wait()
    #self.play(ReplacementTransform(lign4b,lign4c))
    self.play(TransformMatchingTex(lign4b,lign4c),run_time=2)
    self.wait()
    #self.play(ReplacementTransform(lign4c,lign4d))
    self.play(TransformMatchingTex(lign4c,lign4d),run_time=2)
    self.wait()
    self.play(Write(lign5))
    self.wait()
    self.play(Write(lign6a))
    self.wait()
    self.play(TransformMatchingTex(lign6a,lign6b),run_time=2)
    self.wait()
    self.play(TransformMatchingTex(lign6b,lign6c),run_time=2)
    self.wait()
    self.play(TransformMatchingTex(lign6c,lign6d),run_time=2)
    self.wait()
    



```
