```python
%%manim -ql -v WARNING KoordinatsystemPunktAfstanf

class KoordinatsystemPunktAfstanf(Scene):
  def construct(self):
      myTemplate = TexTemplate()
      myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
      myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
      myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
      myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
      myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
      myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
      myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
      self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))

      #tekst = Tex("", tex_template=myTemplate)
      
      #koordinatsystemer

      def polar2c(p):
            return np.array([
                p[0]*np.cos(p[1]),
                p[0]*np.sin(p[1]),
                0
                ])

      PolarSys = PolarPlane(
            azimuth_units="PI radians",
            size=6,
            azimuth_label_font_size=33.6,
            radius_config={"font_size": 33.6},
        ).add_coordinates()
      #self.add(polarplane_pi)
      #self.wait(2)

      KoordSys = NumberPlane(
          x_range=(-6,6,1), 
          y_range=(-6,6,1), 
          x_length=2*PI, 
          y_length=2*PI, 
        ).set_z_index(-1)#.add_coordinates()
      
      KoordSys.generate_target()
      KoordSys.target.prepare_for_nonlinear_transform()
      KoordSys.target.apply_function(lambda p: polar2c(p))



      #decartes
      img_descartes = ImageMobject("drive/MyDrive/img-web/Descartes_manim.jpg").scale(0.25)
      tex_descartes = Tex(r"Ren\,\, ",r"\bf Descartes", tex_template=myTemplate).next_to(img_descartes,DOWN)
      #descartes = VGroup(img_descartes,tex_descartes).move_to(ORIGIN)

      self.play(FadeIn(img_descartes),Write(tex_descartes))
      self.wait()
      self.play(FadeOut(img_descartes),FadeOut(tex_descartes))
      self.wait()
      
      #flyer
      img_fly1 = ImageMobject("drive/MyDrive/img-web/flue").scale(0.05)
      img_fly2 = img_fly1.copy()
      dot_fly1 = Dot(point=KoordSys.coords_to_point(3,2),color=GOLD)
      dot_fly2 = Dot(point=KoordSys.coords_to_point(-2,1),color=GOLD)

      img_fly1.move_to(KoordSys.coords_to_point(3,2))
      img_fly2.move_to(KoordSys.coords_to_point(-2,1))

      #self.add(img_fly1,img_fly2)
      self.play(FadeIn(img_fly1),FadeIn(img_fly2))
      self.wait(2)

      KoordSys_fly = NumberPlane(
          x_range = (-6,6,1), 
          y_range = (-6,6,1), 
          x_length = 2*PI, 
          y_length = 2*PI,
          axis_config = {"stroke_opacity": 0} 
        ).set_z_index(-2)
      #KoordSys_fly.set_opacity(1)
      
      self.play(Create(KoordSys_fly))
      self.wait(3)

      #self.play(Transform(img_fly1,dot_fly1),Transform(img_fly2,dot_fly2))
      self.play(FadeOut(img_fly1),FadeIn(dot_fly1),FadeOut(img_fly2),FadeIn(dot_fly2))
      self.wait()
      #self.play(Create(KoordSys))
      self.add(KoordSys)

      #self.remove(img_fly1,img_fly2)
      #self.wait(2)
      self.remove(KoordSys_fly)

      #aksenavne
      frsteaksen = Tex(r"\bf frste aksen",tex_template=myTemplate).scale(0.5).next_to(KoordSys.coords_to_point(6,0))
      andenaksen = Tex(r"\bf anden aksen",tex_template=myTemplate).scale(0.5).next_to(KoordSys.coords_to_point(0,6))
      x_aksen = Tex(r"x aksen",tex_template=myTemplate).scale(0.4).next_to(frsteaksen,DOWN,buff=SMALL_BUFF).align_to(frsteaksen,LEFT)
      y_aksen = Tex(r"y aksen",tex_template=myTemplate).scale(0.4).next_to(andenaksen,DOWN,buff=SMALL_BUFF).align_to(andenaksen,LEFT)
      abscisse = Tex(r"abscisseaksen",tex_template=myTemplate).scale(0.4).next_to(frsteaksen,DOWN,buff=SMALL_BUFF).align_to(frsteaksen,LEFT)
      ornate = Tex("ornataksen",tex_template=myTemplate).scale(0.4).next_to(andenaksen,DOWN,buff=SMALL_BUFF).align_to(andenaksen,LEFT)


      self.play(Write(frsteaksen))
      self.wait()
      self.play(Write(andenaksen))
      self.wait()

      self.play(Write(x_aksen))
      self.wait()
      self.play(ReplacementTransform(x_aksen,abscisse))
      self.wait()
      self.play(Write(y_aksen))
      self.wait()
      self.play(ReplacementTransform(y_aksen,ornate))
      self.wait()

      self.play(FadeOut(VGroup(abscisse,ornate)))
      self.play(Unwrite(VGroup(frsteaksen,andenaksen)))
      self.wait()
      
      #self.remove(polarplane_pi)
      #self.add(KoordSys)
      #self.play(Transform(KoordSys,PolarSys))
      #self.play(Create(KoordSys))
      self.wait(2)
      self.play(
            MoveToTarget(KoordSys,run_time=4)
        )
      
      self.wait(2)




```

```python
%%manim -ql -v WARNING Polar2Cart

class Polar2Cart(Scene):
    def construct(self):
        def polar2c(p):
            return np.array([
                p[0]*np.cos(p[1]),
                p[0]*np.sin(p[1]),
                0
                ])

        grid = NumberPlane(
            #x_line_frequency=PI/4,
            #y_line_frequency=PI/4,
            x_range=(-4,4,1),
            y_range=(-4,4,1),
            )
        #func = FunctionGraph(lambda x: 0.5*np.sin(5*x)+2,x_range=(-PI,PI))
        #grid.add(func)
        self.add(grid)
        grid.faded_lines[4:9].fade(1)
        grid.faded_lines[12:].fade(1)
        grid.background_lines[4:9].fade(1)
        grid.background_lines[12:].fade(1)
        self.play(Rotating(func,radians=PI,axis=UR,about_point=ORIGIN,run_time=2,rate_func=smooth))
        grid.generate_target()
        grid.target.prepare_for_nonlinear_transform()
        grid.target.apply_function(lambda p: polar2c(p))

        self.play(
            MoveToTarget(grid,run_time=4)
        )
        self.wait(3)
```
