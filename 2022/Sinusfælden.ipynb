```python
%%manim -ql -v WARNING SineTrapTitleIntro

class SineTrapTitleIntro(Scene):
  def construct(self):
    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")

    emne = Tex(r"Trigonometri",tex_template=myTemplate)
    title = Tex(r"\bf N\AA R SINUSRELATIONEN \\ GIVER TO L\O SNINGER", tex_template=myTemplate,tex_environment="flushleft").scale(2).set_color_by_gradient(PURPLE,PINK)

    title.to_corner(DL)
    emne.next_to(title,UP).align_to(title,LEFT)

    self.play(AnimationGroup(*[Write(emne),Write(title)],run_time=2,lag_ratio=0.5))
    self.wait()
    self.play(FadeOut(VGroup(title,emne)))
    self.wait()
```

```python
%%manim -ql -v WARNING SineTrap

class SineTrap(Scene):
  def construct(self):
    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    #myTemplate.add_to_preamble(r"\usepackage[danish]{babel}") #clasher lige nu med en anden pakke
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    emne = Tex(r"Sinus",r"\bf f√¶lden",tex_template=myTemplate).to_corner(UL)
    emne[0].set_color(GOLD)
    
    start_angle = 0
    angleA = ValueTracker(0)

    baseline = DashedLine(start=[-6,-2,0],end=[6,-2,0],dash_length=0.25)
    point = Dot(point=[-3,-2,0]).scale(2)
    line = always_redraw(
        lambda: Line(start=[-3,-2,0],end=[5*np.cos(angleA.get_value())-3,5*np.sin(angleA.get_value())-2,0])
    )
    
    len = ValueTracker(2)
    angleC = ValueTracker(5*PI/3)

    moving_line = always_redraw(
        lambda: Line(
            start=[1.045084971874737,0.938926261462366,0],
            end=[len.get_value()*np.cos(angleC.get_value())+1.045084971874737,len.get_value()*np.sin(angleC.get_value())+0.938926261462366,0]
            )
    )

    A = Tex(r"A",tex_template=myTemplate).next_to(point,LEFT)
    a = Tex(r"a",tex_template=myTemplate).next_to(moving_line.get_center(),RIGHT)#.shift(0.6*LEFT,1.1*DOWN)
    b = Tex(r"b",tex_template=myTemplate).next_to(line,UP)

    self.play(Write(emne))
    self.wait()
    self.play(Create(baseline))
    self.wait()
    self.play(FadeIn(point))
    self.play(Write(A))
    self.wait()
    self.play(Create(line))
    self.play(Write(b))
    self.wait()
    self.play(
        angleA.animate.set_value(PI/5),
        MaintainPositionRelativeTo(b,line),
        run_time=4
    )
    self.wait()
    
    self.play(Create(moving_line))
    self.play(Write(a))
    self.play(
        angleC.animate.set_value(4*PI/3),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=4
    )
    self.play(
        angleC.animate.set_value(5*PI/3),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=4
    )
    self.wait()

    FirstOutcome = MathTex(r"a<h\;\;",r"\texttt{ingen l√∏sning}",tex_template=myTemplate).next_to(emne,DOWN,buff=LARGE_BUFF).align_to(emne,LEFT)

    h = DashedLine(start=[1.045084971874737,0.938926261462366,0],end=[1.045084971874737,-2,0])

    self.play(Create(h))
    self.wait()

    self.play(Write(FirstOutcome))
    self.wait()

    ref_line = Line(
            start=[1.045084971874737,0.938926261462366,0],
            end=[2.938926261462366*np.cos(angleC.get_value())+1.045084971874737,2.938926261462366*np.sin(angleC.get_value())+0.938926261462366,0]
            )

    self.play(
        len.animate.set_value(2.938926261462366),
        a.animate.next_to(ref_line.get_center())
    )
    self.wait()
    self.play(
        angleC.animate.set_value(3*PI/2),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=2
    )

    SecondOutcome = MathTex(r"a=h\;\;",r"\texttt{√©n l√∏sning}",tex_template=myTemplate).next_to(FirstOutcome,DOWN).align_to(emne,LEFT)
    self.play(Write(SecondOutcome))
    self.wait()

    ref_line = Line(
            start=[1.045084971874737,0.938926261462366,0],
            end=[3.5*np.cos(angleC.get_value())+1.045084971874737,3.5*np.sin(angleC.get_value())+0.938926261462366,0]
            )
    
    self.play(
        len.animate.set_value(3.5),
        a.animate.next_to(ref_line.get_center())
    )
    self.wait()
    self.play(
        angleC.animate.set_value(5.286467246),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=2
    )
    self.wait()
    self.play(
        angleC.animate.set_value(4.138310716),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=2
    )
    self.wait()

    ThirdOutcome = MathTex(r"a>h\;\;",r"\texttt{to l√∏sninger}",tex_template=myTemplate).next_to(SecondOutcome,DOWN).align_to(emne,LEFT)
    self.play(Write(ThirdOutcome))
    self.wait()

    ref_line = Line(
            start=[1.045084971874737,0.938926261462366,0],
            end=[6*np.cos(angleC.get_value())+1.045084971874737,6*np.sin(angleC.get_value())+0.938926261462366,0]
            )

    self.play(
        len.animate.set_value(6),
        a.animate.next_to(ref_line.get_center())
    )
    self.wait()
    self.play(
        angleC.animate.set_value(3.653477128),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=2
    )
    self.wait()

    ThirdOutcome2 = MathTex(r"b>a>h\;\;",r"\texttt{to l√∏sninger}",tex_template=myTemplate).next_to(SecondOutcome,DOWN).align_to(emne,LEFT)
    self.play(TransformMatchingShapes(ThirdOutcome,ThirdOutcome2))
    self.wait()

    self.play(
        angleC.animate.set_value(5.771300834),
        MaintainPositionRelativeTo(a,moving_line),
        run_time=2
    )
    self.wait()

    ThirdOutcome = MathTex(r"a>b\;\;",r"\texttt{√©n l√∏sninger}",tex_template=myTemplate).next_to(ThirdOutcome,DOWN).align_to(emne,LEFT)
    self.play(Write(ThirdOutcome))
    self.wait()
    

    self.wait(5)

```

```python
%%manim -ql -v WARNING SineTrapTwoAngles

class SineTrapTwoAngles(Scene):
  def construct(self):
    A_num = MathTex(r"A=36^\circ").to_corner(UL)
    a_num = MathTex(r"a=2").next_to(A_num,DOWN).align_to(A_num,LEFT)
    b_num = MathTex(r"b=5").next_to(a_num,DOWN).align_to(A_num,LEFT)

    SineRel = MathTex(r"\frac{a}{\sin(A)}=\frac{b}{\sin(B)}")
    
    eq1 = MathTex(r"\frac{2}{\sin(36^\circ)}=\frac{5}{\sin(B)").next_to(SineRel,DOWN)
    eq11= MathTex(r"\frac{\sin(36^\circ)}{2}=\frac{\sin(B)}{5}").next_to(SineRel,DOWN)
    eq12= MathTex(r"\frac{\sin(B)}{5}=\frac{\sin(36^\circ)}{2}").next_to(SineRel,DOWN)
    eq13= MathTex(r"\sin(B)=\frac{\sin(36)}{2}\cdot 5").next_to(SineRel,DOWN)
    eq14= MathTex(r"B=\sin^{-1}\Big(\frac{\sin(36)}{2}\cdot 5\Big)").next_to(SineRel,DOWN)
    eq15= MathTex(r"B=\emptyset").next_to(SineRel,DOWN)

    #all = VGroup(SineRel,eq1,eq11,eq12,eq13,eq14).arrange(DOWN)

    self.play(Write(SineRel))
    self.wait()
    self.play(Write(eq1))
    self.wait()
    self.play(TransformMatchingShapes(eq1,eq11))
    self.wait()
    self.play(TransformMatchingShapes(eq11,eq12))
    self.wait()
    self.play(TransformMatchingShapes(eq12,eq13))
    self.wait()
    self.play(TransformMatchingShapes(eq13,eq14))
    self.wait()
    self.play(Transform(eq14,eq15))
    self.wait(5)


```
