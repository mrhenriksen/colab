```python
%%manim -qk -v WARNING OrthoLines

class OrthoLines(Scene):
  def construct(self):
    #self.camera.frame.save_state()
    def alignEQ(eq1name,eq2name):
      a = 0
      b = 0
      pos1 = 0
      pos2 = 0

      for i in (eq1name):
        a += 1
      
      for i in (eq2name):
        b += 1
      for i in range(a):
        x = eq1name[i]
        for y in ("="):
          if x == eq1name.get_part_by_tex(y):
            pos1 = i

      for i in range(b):
        x = eq2name[i]
        for y in ("="):
          if x == eq2name.get_part_by_tex(y):
            pos2 = i
      
      eq2name[pos2].move_to(eq1name[pos1])
      eq2name[0:pos2].next_to(eq1name[pos1],LEFT)
      eq2name[pos2+1:b].next_to(eq2name[pos2],RIGHT)

    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))

# TITLE
    title = Tex(r"\bf ORTOGONALE LINJER",tex_template=myTemplate).scale(1.5).to_edge(UP)
    title2= Tex(r"\bf PARALELLE LINJER", tex_template=myTemplate).scale(1.5).to_edge(UP)

# INITIAL INTRO

    first_eq = MathTex(r"\ell_1\perp\ell_2",tex_template=myTemplate).scale(2)
    second_eq = MathTex(r"a_1\cdot a_2=-1",tex_template=myTemplate).scale(2).next_to(first_eq,RIGHT,buff=LARGE_BUFF)
    third_eq = MathTex(r"\ell_1\nparallel\ell_2",tex_template=myTemplate).scale(2)
    eqs = VGroup(first_eq,second_eq).move_to(ORIGIN)

#    line1 = KoordSys.plot(
#        lambda x: 0.6*x+2.4,
#        x_range = (-20,20),
#        color = PURPLE,
#    )
#    line2 = line1.copy()

# SETUP COORDSYS AND LINES
    KoordSys = NumberPlane(
        x_range = (-5,5,1),
        y_range = (-5,5,1),
        x_length = 5,
        y_length = 5,
    ).set_z_index(-1).add_coordinates()

    func1 = KoordSys.plot(
        lambda x: 0.5*x+0.5,
        x_range = (-20,20),
        color = PURPLE,
    )

    func2 = KoordSys.plot(
        lambda x: -2*x+2.5,
        x_range = (-10,10),
        color = BLUE,
    )

    line1_tex = MathTex(r"\ell_1",tex_template=myTemplate).move_to(KoordSys.coords_to_point(4,2))
    line2_tex = MathTex(r"\ell_2",tex_template=myTemplate).move_to(KoordSys.coords_to_point(2,-2))

    angle1 = Angle.from_three_points(KoordSys.coords_to_point(0,0.5),KoordSys.coords_to_point(0.8,0.9),KoordSys.coords_to_point(0,2.5),elbow=True)

# SETUP FOR ZOOM
    KoordSys2 = NumberPlane(
        x_range = (-5,5,1),
        y_range = (-5,5,1),
        x_length = 15,
        y_length = 15,
    ).set_z_index(-1).add_coordinates().shift(5*LEFT)

    func11 = KoordSys2.plot(
        lambda x: 0.5*x+0.5,
        x_range = (-20,20),
        color = PURPLE,
    ).set_z_index(2)

    func21 = KoordSys2.plot(
        lambda x: -2*x+2.5,
        x_range = (-10,10),
        color = BLUE,
    ).set_z_index(2)

    line11_tex = MathTex(r"\ell_1",tex_template=myTemplate).next_to(KoordSys2.coords_to_point(3,2),DOWN)
    line21_tex = MathTex(r"\ell_2",tex_template=myTemplate).next_to(KoordSys2.coords_to_point(2,-2),RIGHT,buff=LARGE_BUFF)
    si_eq1 = MathTex(r":\; y=a_1\cdot x+b_1",tex_template=myTemplate).next_to(line11_tex,RIGHT)
    si_eq2 = MathTex(r":\; y=a_2\cdot x+b_2",tex_template=myTemplate).next_to(line21_tex,RIGHT)

    angle2 = Angle.from_three_points(KoordSys2.coords_to_point(0,0.5),KoordSys2.coords_to_point(0.8,0.9),KoordSys2.coords_to_point(0,2.5),elbow=True,color=GREEN)


# SETUP FOR THE VISUAL SLOPES
    point1 = KoordSys2.coords_to_point(0.8,0.9)
    point2 = KoordSys2.coords_to_point(1.8,1.4)
    point3 = KoordSys2.coords_to_point(1.8,0.9)
    point4 = KoordSys2.coords_to_point(1.8,-1.1)

    slope1 = Polygon(point1,point2,point3,color=GOLD).set_z_index(-1)
    slope2 = Polygon(point1,point3,point4,color=GOLD).set_z_index(-1)

    run = Tex(r"1",tex_template=myTemplate).next_to(KoordSys2.coords_to_point(1.3,0.9),DOWN)
    rise1 = MathTex(r"a_1", tex_template=myTemplate).next_to(KoordSys2.coords_to_point(1.8,1.15),RIGHT)
    rise2 = MathTex(r"a_2", tex_template=myTemplate).next_to(KoordSys2.coords_to_point(1.8,0.4),RIGHT)

    slope11 = slope1.copy()#VGroup(slope1.copy(),run.copy(),rise1)
    slope21 = slope2.copy()#VGroup(slope2.copy(),run.copy(),rise2)
    
    run11 = run.copy().move_to([5.375,2.4,0])
    rise11 = rise1.copy().move_to([6.4,3.1,0])
    run21 = run.copy().move_to([6.3,1,0])
    rise21 = rise2.copy().move_to([4.8,-0.2,0])

    rise22 = MathTex(r"-a_2",tex_template=myTemplate).move_to([4.8,-0.2,0])

# EQUATIONS
    eq1 = MathTex(
        r"{",#0
        r"a_1",#1 - rise11
        r"\over",#2
        r"1",#3 - run11
        r"}",#4
        r"=",#5
        r"{",#6
        r"1",#7 - run21
        r"\over",#8
        r"-a_2",#9 - rise21
        r"}",#10
        tex_template=myTemplate,
        ).scale(1.5).to_corner(DR).shift(0.75*UP)
    eq2 = MathTex(
        r"{",#0
        r"a_1",#1 - rise11
        r"\over",#2
        r"1",#3 - run11
        r"}",#4
        r"=",#5
        r"-",#6
        r"{",#7
        r"1",#8 - run21
        r"\over",#9
        r"a_2",#10 - rise21
        r"}",#11
        tex_template=myTemplate,
    ).scale(1.5)
    eq3 = MathTex(
        r"a_1",#0
        r"=",#1
        r"-",#2
        r"{",#3
        r"1",#4 - run21
        r"\over",#5
        r"a_2",#6 - rise21
        r"}",#7
        tex_template=myTemplate,
    ).scale(1.5)
    eq4 = MathTex(
        r"a_1",#0
        r"\cdot",#1
        r"a_2",#2
        r"=",#3
        r"-1",#4
        tex_template=myTemplate,
    ).scale(1.5)
    alignEQ(eq1,eq2)
    alignEQ(eq1,eq3)
    alignEQ(eq1,eq4)

    box = SurroundingRectangle(eq4,corner_radius=0.2,buff=MED_LARGE_BUFF,color=PINK,fill_opacity=0.1).set_z_index(-1)

# CREATE TITLE
    title.set_z_index(2)
    self.play(Write(title))
    self.wait()

# INTRODUCTION
    self.play(Write(first_eq))
    self.wait()
    self.play(Write(second_eq))
    self.wait(2)
    self.play(Unwrite(second_eq))
    self.play(Unwrite(first_eq))
    self.wait()

# CREATE COORDSYS AND LINES
    #self.add(KoordSys,func1,func2,line1_tex,line2_tex)
    self.play(Create(func1))
    self.play(Write(line1_tex))
    self.wait()
    self.play(ReplacementTransform(func1.copy(),func2))
    self.play(Write(line2_tex))
    self.wait()
    self.play(Create(angle1))
    self.wait()
    self.play(Create(KoordSys))
    self.wait()

# ZOOM TO GET A BETTER VIEW
    self.play(
        Transform(KoordSys,KoordSys2),
        Transform(func1,func11),
        Transform(func2,func21),
        Transform(line1_tex,line11_tex),
        Transform(line2_tex,line21_tex),
        Transform(angle1,angle2),
        )    
    self.wait()
    #self.play(self.camera.frame.animate.scale(0.35).move_to(KoordSys.coords_to_point(4,0)))

# ADDING VISUAL SLOPES
    #self.add(slope1,slope2,run,rise1,rise2)
    self.play(Write(si_eq1))
    self.wait()
    self.play(Write(si_eq2))
    self.wait()
    self.play(Create(slope1))
    self.play(Write(run),Write(rise1))
    self.wait()
    self.play(Create(slope2))
    self.play(Write(rise2))
    self.wait()
    self.play(
        slope11.animate.to_corner(UR).shift(0.5*LEFT),
        slope21.animate.to_corner(UR).shift(DOWN+1.3*LEFT).rotate(-PI/2),
        ReplacementTransform(run.copy(),run11),
        ReplacementTransform(run.copy(),run21),
        ReplacementTransform(rise2.copy(),rise21),
        ReplacementTransform(rise1.copy(),rise11),
        )
    self.wait()
    self.play(ReplacementTransform(rise21,rise22))
    self.wait()
    #self.add(eq1)
    self.play(ReplacementTransform(rise11.copy(),eq1[1]),ReplacementTransform(run11.copy(),eq1[3]),Create(eq1[2]))
    self.wait()
    self.play(ReplacementTransform(rise21.copy(),eq1[9]),ReplacementTransform(run21.copy(),eq1[7]),Create(eq1[8]))
    self.wait()
    self.play(Create(eq1[5]))
    self.wait()
    self.play(ReplacementTransform(eq1[0:5],eq3[0:1],run_time=2))
    self.wait()
    self.play(ReplacementTransform(eq1[6:10],eq3[2:8]),run_time=2)
    self.wait()
    self.play(TransformMatchingTex(eq3,eq4,path_arc=PI/2,run_time=2))
    self.wait()
    self.play(Create(box))

    self.wait(2)
    self.play(*[FadeOut(mob) for mob in self.mobjects])    
    self.wait(5)
```

```python
%%manim -qk -v WARNING OrthoLinesIntro

class OrthoLinesIntro(Scene):
  def construct(self):
    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))

# TITLE
    title = Tex(r"\bf ORTOGONALE LINJER",tex_template=myTemplate).scale(1).to_corner(UR)
    title2= Tex(r"\bf PARALLELLE LINJER", tex_template=myTemplate).scale(1).to_corner(UL)

# INITIAL INTRO
    first_eq = MathTex(r"\ell_1\perp\ell_2",tex_template=myTemplate).scale(1.5).next_to(title,DOWN,buff=MED_LARGE_BUFF)
    second_eq = MathTex(r"a_1\cdot a_2=-1",tex_template=myTemplate).scale(1.5).next_to(first_eq,DOWN)
    third_eq = MathTex(r"\ell_1\|\ell_2",tex_template=myTemplate).scale(1.5).next_to(title2,DOWN,buff=MED_LARGE_BUFF)
    forth_eq = MathTex(r"a_1=a_2",tex_template=myTemplate).scale(1.5).next_to(third_eq,DOWN)
    fifth_eq = MathTex(r"\ell_1\nparallel\ell_2",tex_template=myTemplate).scale(1.5).move_to(third_eq)
    #eqs = VGroup(first_eq,second_eq).move_to(ORIGIN)

    KoordSys = NumberPlane(
        x_range = (-5,5,1),
        y_range = (-5,5,1),
        x_length = 5,
        y_length = 5,
    ).set_z_index(-1).add_coordinates()

    line1 = KoordSys.plot(
        lambda x: 0.6*x-2.4,
        x_range = (-20,20),
        color = PURPLE,
    )
    line2 = KoordSys.plot(
        lambda x: 0.6*x-5.4,
        x_range = (-20,20),
        color = PURPLE,
    )

    angle = Angle.from_three_points(KoordSys.coords_to_point(0,-2.4),KoordSys.coords_to_point(-45/34,-3.19411765),KoordSys.coords_to_point(-3.24,0),elbow=True)

    dot1 = Dot(KoordSys.coords_to_point(-10,-8.4))
    dot2 = Dot(KoordSys.coords_to_point(-10,-11.4))
    
    #self.add(title,title2,line1,line2,third_eq,forth_eq,dot1)
    self.play(Succession(Write(title2),Write(title),lag_ratio=0.5))
    self.wait()
    self.play(Create(line1),Create(line2),title.animate.set_color(DARK_GRAY))
    self.play(Write(third_eq))
    self.wait()
    self.play(Write(forth_eq))
    self.wait()
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-7,-8.4)),
              dot2.animate.move_to(KoordSys.coords_to_point(-7,-11.4)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-7,-6.6)),
              dot2.animate.move_to(KoordSys.coords_to_point(-7,-9.6)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-4,-6.6)),
              dot2.animate.move_to(KoordSys.coords_to_point(-4,-9.6)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-4,-4.8)),
              dot2.animate.move_to(KoordSys.coords_to_point(-4,-7.8)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-1,-4.8)),
              dot2.animate.move_to(KoordSys.coords_to_point(-1,-7.8)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(-1,-3.0)),
              dot2.animate.move_to(KoordSys.coords_to_point(-1,-6.0)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(2,-3.0)),
              dot2.animate.move_to(KoordSys.coords_to_point(2,-6.0)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(2,-1.2)),
              dot2.animate.move_to(KoordSys.coords_to_point(2,-4.2)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(5,-1.2)),
              dot2.animate.move_to(KoordSys.coords_to_point(5,-4.2)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(5,0.6)),
              dot2.animate.move_to(KoordSys.coords_to_point(5,-2.4)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(8,0.6)),
              dot2.animate.move_to(KoordSys.coords_to_point(8,-2.4)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(8,2.4)),
              dot2.animate.move_to(KoordSys.coords_to_point(8,-0.6)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(11,2.4)),
              dot2.animate.move_to(KoordSys.coords_to_point(11,-0.6)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(11,4.2)),
              dot2.animate.move_to(KoordSys.coords_to_point(11,1.2)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(14,4.2)),
              dot2.animate.move_to(KoordSys.coords_to_point(14,1.2)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(14,6.0)),
              dot2.animate.move_to(KoordSys.coords_to_point(14,3.0)),
              run_time=0.5,
              )
    self.play(dot1.animate.move_to(KoordSys.coords_to_point(17,6.0)),
              dot2.animate.move_to(KoordSys.coords_to_point(17,3.0)),
              run_time=0.5,
              )
    
    self.play(FadeOut(forth_eq),
              ReplacementTransform(third_eq,fifth_eq),
              title2.animate.set_color(DARK_GRAY),
              line2.animate.rotate(PI/4).set_color_by_gradient(PURPLE,BLUE))
    
    self.wait()

    self.play(FadeOut(fifth_eq),
              title.animate.set_color(WHITE),
              line2.animate.rotate(PI/4).set_color(BLUE)
              )
    self.play(Create(angle),Succession(Write(first_eq),Write(second_eq),lag_ratio=1),
             )
    self.wait(2)
    self.play(*[FadeOut(mob) for mob in self.mobjects])
    self.wait(3)

```
