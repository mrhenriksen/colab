```python
%%manim -ql -v WARNING Tvrvektor

class TvrvektorDef(Scene):
  def construct(self):

    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))
    
    # TITLE
    title = Tex(r"\bf TVRVEKTOR",tex_template=myTemplate).scale(1.5).to_corner(UL)
    
    # DEFINITION
    definition_title = Tex(r"\bf DEFINITION",tex_template=myTemplate).set_color(MAROON).move_to(UL)
    definition1 = Tex(r"En ",r"\bf tvrvektor ",r"er en ",r"vektor\\ ",tex_template=myTemplate).next_to(definition_title,DOWN).align_to(definition_title,LEFT)
    definition2 = Tex(r"der er ",r"drejet ",r"90$^\circ$ ",r"\bf mod uret",tex_template=myTemplate).next_to(definition_title,DOWN).align_to(definition_title,LEFT).next_to(definition1,DOWN).align_to(definition1,LEFT)

    definition1[1].set_color(PURPLE_A)
    definition1[3].set_color(PURPLE_B)
    definition2[1].set_color(PURPLE_C)
    definition2[2].set_color(PURPLE_D)
    definition2[3].set_color(PURPLE_E)

    self.play(Write(title))
    self.wait()
    self.play(Write(definition_title))
    self.play(Write(definition1))
    self.play(Write(definition2))
    self.wait()

class Tvrvektor(Scene):
  def construct(self):

    myTemplate = TexTemplate()
    myTemplate.add_to_preamble(r"\usepackage[utf8]{inputenc}",prepend=True)
    myTemplate.add_to_preamble(r"\usepackage[scaled=0.95]{helvet}")
    myTemplate.add_to_preamble(r"\renewcommand\familydefault{\sfdefault}")
    myTemplate.add_to_preamble(r"\usepackage[T1]{fontenc}")
    myTemplate.add_to_preamble(r"\usepackage[helvet]{sfmath}")
    myTemplate.add_to_preamble(r"\usepackage{amsmath,bm}")
    myTemplate.add_to_preamble(r"\usepackage{upgreek}")
    
    self.camera.background_color=rgb_to_color(hex_to_rgb("#0A2C33"))

    # TITLE
    title = Tex(r"\bf TVRVEKTOR",tex_template=myTemplate).scale(1.5).to_corner(UL).set_z_index(3)
    self.add(title)

    def MyVector(koords,start,**kwargs):
      x_start = start[0]
      y_start = start[1]
      x_koord = koords[0]
      y_koord = koords[1]
      myvector = Vector(koords).move_to([x_start+(x_koord/2),y_start+(y_koord/2),0])
      return myvector

    def vekkoord(vektor):
      vect_start = vektor.get_start()
      vect_slut = vektor.get_end()
      vect_x = [vect_slut[0]-vect_start[0]]
      vect_y = [vect_slut[1]-vect_start[1]]

      label = Matrix(np.round([vect_x,vect_y]).astype(int),left_bracket="(",right_bracket=")")
      if np.round([vect_x]).astype(int) > 0:
        label.next_to(vect_slut,RIGHT)
      else:
        label.next_to(vect_slut,LEFT)
      return label
    
    plane = NumberPlane()
    start_point = [0,0,0]
    vektor = MyVector([1,3],start_point)

    pil = CurvedArrow([1,3.1,0],[0,3.3,0],radius=3.3).set_color(GOLD)

    #tvrvektor = vektor.rotate(angle=PI/2,about_point=start_point)
    vinkel = Angle.from_three_points([1,3,0],[0,0,0],[-3,1,0],elbow=True).set_color(GREEN_D)

    ax = DashedLine([0,0,0],[1,0,0]).set_color(YELLOW)
    ay = DashedLine([1,0,0],[1,3,0]).set_color(YELLOW)
    tax = DashedLine([0,0,0],[-3,0,0]).set_color(YELLOW)
    tay = DashedLine([-3,0,0],[-3,1,0]).set_color(YELLOW)
    axy = VGroup(ax,ay)
    taxy = VGroup(tax,tay)

    ax_label = MathTex(r"a_x",tex_template=myTemplate).next_to(ax,DOWN)
    ay_label = MathTex(r"a_y",tex_template=myTemplate).next_to(ay,RIGHT)
    tax1_label = MathTex(r"a_y",tex_template=myTemplate).next_to(tax,DOWN)
    tax2_label = MathTex(r"-a_y",tex_template=myTemplate).next_to(tax,DOWN).align_to(tax1_label,RIGHT)
    tay_label = MathTex(r"a_x",tex_template=myTemplate).next_to(tay,LEFT)

    vinkel1 = Angle.from_three_points([1,0,0],[0,0,0],[1,3,0]).set_color(GREEN_D)
    vinkel21 = Angle.from_three_points([0,0,0],[1,3,0],[1,0,0],radius=0.7).set_color(GREEN_D)
    vinkel22 = Angle.from_three_points([0,0,0],[1,3,0],[1,0,0],radius=0.7+0.1).set_color(GREEN_D)
    vinkel2 = VGroup(vinkel21,vinkel22)
    vinkel31 = Angle.from_three_points([-3,1,0],[0,0,0],[-3,0,0],radius=0.7).set_color(GREEN_D)
    vinkel32 = Angle.from_three_points([-3,1,0],[0,0,0],[-3,0,0],radius=0.7+0.1).set_color(GREEN_D)
    vinkel3 = VGroup(vinkel31,vinkel32)
    vinkel4 = Angle.from_three_points([-3,0,0],[-3,1,0],[0,0,0],radius=0.4).set_color(GREEN_D)
    #vinkel42 = Angle.from_three_points([-3,0,0],[-3,1,0],[0,0,0],radius=0.4+0.1).set_color(GREEN_D)
    #vinkel4 = VGroup(vinkel41,vinkel42)

    vinkel8 = Angle.from_three_points([1,3,0],[1,0,0],[0,0,0],elbow=True).set_color(GREEN_D)


    vinkel5 = vinkel1.copy().move_to([3,2,0])
    plus = MathTex(r"+",tex_template=myTemplate).next_to(vinkel5,RIGHT)
    vinkel6 = vinkel2.copy().rotate(angle=PI/2 + np.arctan(1/3)).next_to(plus,RIGHT)
    plus2 = MathTex(r"+",tex_template=myTemplate).next_to(vinkel6,RIGHT)
    vinkel7 = vinkel.copy().rotate(angle=-np.arctan(3)).next_to(plus2,RIGHT)
    ligmed = MathTex(r"=",r"180^\circ",tex_template=myTemplate).next_to(vinkel7,RIGHT)

    ligedannede = Tex(r"\bf ligedannede").move_to(plane.c2p(3,1))

    koord_label_pos = vekkoord(vektor)

    self.add(plane)
    self.wait()
    self.play(Create(vektor))
    self.wait()
    self.play(FadeIn(pil))
    self.wait()
    self.play(FadeOut(pil))
    self.wait()
    self.play(Rotate(vektor.copy(),angle=PI/2,about_point=start_point))
    self.wait()
    self.play(Create(vinkel))
    self.wait()
    #self.play(Uncreate(vinkel))
    self.play(vinkel.animate.set_color(DARK_GRAY))
    self.wait()
    self.play(Create(axy))
    self.wait()
    self.play(Create(vinkel8))
    self.wait()
    self.play(Write(ax_label),Write(ay_label))
    self.wait()
    self.play(Create(taxy))
    #self.add(ax_label,ay_label,tax1_label,tay_label)
    self.wait()
    self.play(Create(vinkel1))
    self.play(Create(vinkel2))

    self.wait()
    self.play(ReplacementTransform(vinkel1.copy(),vinkel5))
    self.play(Create(plus))
    self.play(ReplacementTransform(vinkel2.copy(),vinkel6))
    self.play(Create(plus2))
    self.play(ReplacementTransform(vinkel8,vinkel7))
    self.play(Create(ligmed))
    self.wait()

    self.play(ReplacementTransform(vinkel5,vinkel1))
    self.wait()
    #self.play(Create(vinkel))
    self.play(ReplacementTransform(vinkel7,vinkel),vinkel.animate.set_color(GREEN_D))
    #self.add(vinkel)
    self.wait()
    self.play(ReplacementTransform(vinkel6,vinkel3),FadeOut(plus),FadeOut(plus2),FadeOut(ligmed))
    self.wait()
    self.play(ReplacementTransform(vinkel1.copy(),vinkel4))
    self.wait()

    self.play(FadeIn(ligedannede))
    self.wait()
    self.play(FadeOut(ligedannede))
    self.wait()

    self.play(Write(tax1_label),Write(tay_label))
    self.wait()
    self.play(TransformMatchingTex(tax1_label,tax2_label))

    vektor_formel_navn = MathTex(r"\vec{a}=",tex_template=myTemplate)
    vektor_formel_koord = MobjectMatrix([MathTex(r"a_x",tex_template=myTemplate),MathTex(r"ay",tex_template=myTemplate)],left_bracket="(",right_bracket=")").next_to(vektor_formel_navn,RIGHT)
    vektor_formel=VGroup(vektor_formel_navn,vektor_formel_koord).scale(0.75).next_to([2,1.5,0],RIGHT)
    tvrvektor_formel_navn = MathTex(r"\widehat{\vec{a}}=",tex_template=myTemplate)
    tvrvektor_formel_koord = MobjectMatrix([MathTex(r"-a_y",tex_template=myTemplate),MathTex(r"a_x",tex_template=myTemplate)],left_bracket="(",right_bracket=")").next_to(tvrvektor_formel_navn,RIGHT)
    tvrvektor_formel = VGroup(tvrvektor_formel_navn,tvrvektor_formel_koord).scale(0.75).next_to([-4,0.5,0],LEFT)

    self.play(Write(vektor_formel))
    self.wait()
    self.play(Write(tvrvektor_formel))
    self.wait()
    self.play(vektor_formel.animate.to_corner(UR))
    self.play(tvrvektor_formel.animate.next_to(vektor_formel,DOWN).align_to(vektor_formel,RIGHT))    
    #self.play(Create(vinkel3))
    #self.play(Create(vinkel4))

#    self.play(Write(koord_label_pos))
    self.wait()
    self.wait()
```
